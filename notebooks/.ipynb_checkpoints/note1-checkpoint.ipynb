```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import from_json, col
from pyspark.sql.types import StructType, StringType, FloatType

# 1. Créer la SparkSession
spark = SparkSession.builder \
    .appName("RealTimeStockSentiment") \
    .config("spark.jars.packages", "org.apache.spark:spark-sql-kafka-0-10_2.12:3.1.1") \
    .getOrCreate()

# 2. Schéma des messages JSON attendus
schema = StructType() \
    .add("Date", StringType()) \
    .add("Tweet", StringType()) \
    .add("Stock Name", StringType()) \
    .add("Company Name", StringType()) \
    .add("Open", FloatType()) \
    .add("High", FloatType()) \
    .add("Low", FloatType()) \
    .add("Close", FloatType()) \
    .add("Volume", FloatType()) \
    .add("Ticker", StringType())

# 3. Lire depuis Kafka
df_raw = spark.readStream \
    .format("kafka") \
    .option("kafka.bootstrap.servers", "kafka:29092") \
    .option("subscribe", "stock_sentiment") \
    .option("startingOffsets", "earliest") \
    .load()

# 4. Convertir la valeur binaire en string
df_value = df_raw.selectExpr("CAST(value AS STRING)")

# 5. Afficher les données brutes Kafka
df_value.writeStream \
    .outputMode("append") \
    .format("console") \
    .option("truncate", False) \
    .start()

# 6. Parser les données JSON
df_parsed = df_value \
    .select(from_json(col("value"), schema).alias("data")) \
    .select("data.*")

# 7. Afficher les données parsées
df_parsed.writeStream \
    .outputMode("append") \
    .format("console") \
    .option("truncate", False) \
    .start() \
    .awaitTermination()
```
