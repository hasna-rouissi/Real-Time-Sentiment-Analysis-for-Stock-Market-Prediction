```python
# importer les bibliotheques necessaires
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, udf
from pyspark.sql.types import StringType
from pyspark.ml.feature import Tokenizer, StopWordsRemover, HashingTF, IDF
from pyspark.ml.classification import LogisticRegression
from pyspark.ml import Pipeline
from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from wordcloud import WordCloud
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import json
```

```python
# sauvegardes la session Spark
spark = SparkSession.builder.appName("SentimentAnalysis").getOrCreate()
```

```python

# Charger sans header, puis renommer
df = spark.read.csv("stock_tweets.csv", 
                    header=True, 
                    inferSchema=True, 
                    multiLine=True,  # Important pour les tweets multi-lignes
                    quote='"',
                    escape="\\",
                    encoding="utf-8")
# Affichage de quelques lignes
df.show(truncate=False)
```

```python
from pyspark.sql.functions import regexp_replace

# Nettoyer les caractÃ¨res non ASCII
df = df.withColumn("Tweet", regexp_replace("Tweet", "[^\\x00-\\x7F]", ""))
df
```

```python
# Nettoyage des caractÃ¨res spÃ©ciaux, liens, mentions, hashtags, etc.
df_clean = df.withColumn("clean_tweet", regexp_replace(col("Tweet"), r"http\S+|www\S+|https\S+", "")) \
             .withColumn("clean_tweet", regexp_replace(col("clean_tweet"), r"@\w+", "")) \
             .withColumn("clean_tweet", regexp_replace(col("clean_tweet"), r"#", "")) \
             .withColumn("clean_tweet", regexp_replace(col("clean_tweet"), r"[^a-zA-Z\s]", "")) \
             .withColumn("clean_tweet", regexp_replace(col("clean_tweet"), r"\s+", " ")) \
             .withColumn("clean_tweet", col("clean_tweet").alias("clean_tweet"))

df_clean.select("Tweet", "clean_tweet").show(5, truncate=False)
```

```python
from textblob import TextBlob
from pyspark.sql.functions import udf, col
from pyspark.sql.types import IntegerType

# Fonction pour dÃ©tecter le sentiment (0 = nÃ©gatif, 4 = positif)
def get_sentiment(text):
    if not text:
        return 0
    analysis = TextBlob(text)
    return 4 if analysis.sentiment.polarity > 0 else 0

# Enregistrer comme une UDF Spark
sentiment_udf = udf(get_sentiment, IntegerType())

# Ajouter une colonne 'target' en analysant 'clean_tweet'
df_clean = df_clean.withColumn("target", sentiment_udf(col("clean_tweet")))
```

```python
df_clean.show(truncate=False)
```

```python
from pyspark.ml.feature import Tokenizer, StopWordsRemover, HashingTF, IDF

# Tokenization
tokenizer = Tokenizer(inputCol="clean_tweet", outputCol="words")

# Suppression des stop words
remover = StopWordsRemover(inputCol="words", outputCol="filtered")

# TF
hashingTF = HashingTF(inputCol="filtered", outputCol="rawFeatures", numFeatures=10000)

# TF-IDF
idf = IDF(inputCol="rawFeatures", outputCol="features")
```

```python
from pyspark.ml.feature import StringIndexer

indexer = StringIndexer(inputCol="target", outputCol="label")
```

```python
from pyspark.sql.functions import col

df_clean = df_clean.filter(col("clean_tweet").isNotNull())
```

```python
lr = LogisticRegression(featuresCol="features", labelCol="label")# Applique un modÃ¨le dâ€™apprentissage supervisÃ© pour prÃ©dire le label (positif ou nÃ©gatif)
rf = RandomForestClassifier(featuresCol="features", labelCol="label", numTrees=50)

pipeline_lr = Pipeline(stages=[tokenizer, remover, hashingTF, idf, indexer, lr])

pipeline_rf = Pipeline(stages=[tokenizer, remover, hashingTF, idf, indexer, rf])
```

```python
# entraÃ®ner le modÃ¨le
model_lr = pipeline_lr.fit(df_clean)
model_rf = pipeline_rf.fit(df_clean)
```

```python
#prÃ©dire sur de nouveaux tweets

pred_lr = model_lr.transform(df_clean)
pred_rf = model_rf.transform(df_clean)

pred_lr.select("clean_tweet", "label", "prediction").show(10)
pred_rf.select("clean_tweet", "label", "prediction").show(10)
```

```python
evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName="accuracy")

accuracy_lr = evaluator.evaluate(pred_lr)
accuracy_rf = evaluator.evaluate(pred_rf)

print(f"Accuracy Logistic Regression: {accuracy_lr}")
print(f"Accuracy Random Forest     : {accuracy_rf}")
```
